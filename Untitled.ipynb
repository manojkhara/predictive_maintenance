# Variable Explaination

Now that we have the data imported into a Jupiter Notebook, we can explore it. Here is metadata explaining all of the fields in the data set.

### Independent Variable

1) ID — ID field that represents a specific machine.

2) DATE — The date of the observation.

3) REGION_CLUSTER — a field that represents the region in which the machine resides.

4) MAINTENANCE_VENDOR — a field that represents the company that provides maintenance and service to the machine.

5) MANUFACTURER — the company that manufactured the equipment in question.

6) WELL_GROUP — a field representing the type of machine.

7) S5 — A Sensor Value.

8) S8 — A Sensor Value.

9) S13 — A Sensor Value.

10) S15 — A Sensor Value.

11) S16 — A Sensor Value.

12) S17 — A Sensor Value.

13) S18 — A Sensor Value.

14) S19 — A Sensor Value.

15) EQUIPMENT_AGE — Age of the machine, in days.

### Dependent Variable

16) EQUIPMENT_FAILURE — A ‘1’ means that the equipment failed. A ‘0’ means the equipment did not fail.
